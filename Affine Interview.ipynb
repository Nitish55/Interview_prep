```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"city":["l","j","n","v","b","t","my","ha","bh","gw","in","ja"],
                "state":["U.P","U.P","U.P","U.P","K.A","K.A","K.A","K.A","M.P","M.P","M.P","M.P",],
                "revenue":[1000,3000,3450,900,34560,1200,9800,5660,3000,500,2300,1500]})
df
```

```python
ab=pd.DataFrame(df.groupby("state")["revenue"].mean().reset_index())
ab
```

```python
df=pd.merge(df,ab,on="state",how="inner")
df["new_col"]=np.where(df["revenue_x"]>df["revenue_y"],"above_avg","below_avg")
df
```
